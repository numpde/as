```python
import topology
import networkx as nx

for seed in range(0, 1000) :
	G = nx.fast_gnp_random_graph(20, 0.5, seed=seed)
	B = topology.betti(nx.find_cliques(G))
	print(seed, B)
	if any((b == 0) for b in B) : continue

#G = nx.fast_gnp_random_graph(20, 0.5, seed=157)
#B = topology.betti(nx.find_cliques(G))
#print(B)
```
